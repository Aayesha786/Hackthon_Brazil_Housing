# Identify the type of data in each column (categorical or numerical)?

Seperate out the categorical columns from the numerical types



```python
def separate(housing_data):
    categorical = []
    continuous = []
    for column in housing_data.columns:
        if housing_data[column].nunique() < 40:
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical, continuous

categorical, continuous = separate(housing_data)

from tabulate import tabulate

print(tabulate({"Categorical": categorical, "Continuous": continuous}, headers=["Categorical", "Continuous"]))

        
```

# 4) What will be the Average Rent by City?

# 8) How does the average fire insurance impact the rent amount city-wise?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of cities to analyze
cities_to_analyze = ['Porto Alegre', 'Campinas', 'Belo Horizonte']

# Create subplots for city-wise analysis
fig, axes = plt.subplots(len(cities_to_analyze), 1, figsize=(12, 18), sharex=True)

for i, city in enumerate(cities_to_analyze):
    # Filter data for the current city
    city_data = filtered_data[filtered_data['city'] == city]
    
    # Define the categories for rent amounts
    bins = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
    
    # Create a new column to categorize rent amounts
    city_data['rent_amount_category'] = pd.cut(city_data['rent amount (R$)'], bins=bins)
    
    # Calculate the average fire insurance for each category in the current city
    average_fire_insurance = city_data.groupby('rent_amount_category')['fire insurance (R$)'].mean().reset_index()
    
    # Create a bar plot for the current city
    sns.barplot(data=average_fire_insurance, x='rent_amount_category', y='fire insurance (R$)', ax=axes[i])
    axes[i].set_title(f'Average Fire Insurance by Rent Amount Category in {city}')
    axes[i].set_xlabel('Rent Amount Category (R$)')
    axes[i].set_ylabel('Average Fire Insurance (R$)')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust subplot spacing
plt.tight_layout()
plt.show()
```

# Yes, properties that have higher average fire insurance impact the rent, because the higher the rent, the higher the fire insurance. Rent with 10k fire insurance is 140. 

# 9) How does average property tax impact the average rent amount city-wise?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of cities to analyze
cities_to_analyze = ['Porto Alegre', 'Campinas', 'Belo Horizonte']

# Create subplots for city-wise analysis
fig, axes = plt.subplots(len(cities_to_analyze), 1, figsize=(12, 18), sharex=True)

for i, city in enumerate(cities_to_analyze):
    # Filter data for the current city
    city_data = filtered_data[filtered_data['city'] == city]
    
    # Create bins for rent amount
    bins = [0, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 20000, float('inf')]
    labels = ['0-2000', '2000-4000', '4000-6000', '6000-8000', '8000-10000', '10000-12000',
              '12000-14000', '14000-16000', '16000-18000', '18000-20000', '20000+']

    city_data['rent_range'] = pd.cut(city_data['rent amount (R$)'], bins=bins, labels=labels)
    
    # Create a bar plot for the current city
    sns.barplot(data=city_data, x='rent_range', y='property tax (R$)', ci=None, ax=axes[i])
    axes[i].set_title(f'Average Property Tax by Rent Amount Range in {city}')
    axes[i].set_xlabel('Rent Amount Range (R$)')
    axes[i].set_ylabel('Average Property Tax (R$)')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust subplot spacing
plt.tight_layout()
plt.show()
```

# Yes, properties that have a higher average property tax impact the rent, because the higher the rent, the higher the property tax. Rent with 10k to 15k property tax is 1200. Lower the rent, lower the tax. Campinas and Porto collect lower rent, which means lower average property tax.

# 12) Does the Area impact the rental value?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of cities to analyze
cities_to_analyze = ['Porto Alegre', 'Campinas', 'Belo Horizonte']

# Create subplots for city-wise analysis
fig, axes = plt.subplots(len(cities_to_analyze), 1, figsize=(12, 18), sharex=True)

for i, city in enumerate(cities_to_analyze):
    # Filter data for the current city
    city_data = filtered_data[filtered_data['city'] == city]
    
    # Group the data by property area range (e.g., bins) and calculate the median rent amount for each range
    bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, float('inf')]
    labels = ['0-50', '50-100', '100-150', '150-200', '200-250', '250-300', '300-350', '350-400', '400-450', '450-500', '500+']
    city_data['area_range'] = pd.cut(city_data['area'], bins=bins, labels=labels)
    area_rent_median = city_data.groupby('area_range')['rent amount (R$)'].median().reset_index()
    
    # Create a grouped bar chart for the current city
    sns.barplot(data=area_rent_median, x='area_range', y='rent amount (R$)', ci=False, ax=axes[i])
    axes[i].set_title(f'Impact of Area on Rent Amount in {city}')
    axes[i].set_xlabel('Property Area Range (sqm)')
    axes[i].set_ylabel('Median Rent Amount (R$)')
    
    # Rotate x-axis labels for better visibility
    axes[i].tick_params(axis='x', rotation=45)
    
# Adjust subplot spacing
plt.tight_layout()
plt.show()

```

# Yes, properties that have a larger area are more expensive. Because the larger the area, the more the rent is impacted. Campinas and Porto, with smaller areas, collect less rent compared to Belo Horizonte

# 13) Which City is Suitable for Bachelors by Average Rent ? 

```python
# Define the list of cities to analyze
cities_to_analyze = ['Porto Alegre', 'Campinas', 'Belo Horizonte']

# Create dictionaries to store average rent prices for each category and city
avg_rent_bachelor_by_city = {}
avg_rent_mid_sized_by_city = {}
avg_rent_large_by_city = {}

# Loop through each city
for city in cities_to_analyze:
    # Filter data for the current city
    city_data = filtered_data[filtered_data['city'] == city]
    
    # Filter for bachelors (example filters, you can customize these)
    bachelor_properties = city_data[(city_data['rooms'] == 1) & (city_data['bathroom'] == 1)]
    avg_rent_bachelor_by_city[city] = bachelor_properties['rent amount (R$)'].mean()
    
    # Filter for mid-sized families (example filters)
    mid_sized_properties = city_data[(city_data['rooms'] >= 2) & (city_data['rooms'] <= 3) & (city_data['bathroom'] >= 2)]
    avg_rent_mid_sized_by_city[city] = mid_sized_properties['rent amount (R$)'].mean()
    
    # Filter for large families (example filters)
    large_properties = city_data[(city_data['rooms'] > 3) & (city_data['parking spaces'] >= 2)]
    avg_rent_large_by_city[city] = large_properties['rent amount (R$)'].mean()

# Print average rent prices for each category and city
for city in cities_to_analyze:
    print(f"City: {city}")
    print(f"Average Rent for Bachelors: {avg_rent_bachelor_by_city[city]}")
    print(f"Average Rent for Mid-sized Families: {avg_rent_mid_sized_by_city[city]}")
    print(f"Average Rent for Large Families: {avg_rent_large_by_city[city]}\n")
```
